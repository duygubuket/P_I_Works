# Code to fill the missing data (impute) 

```python
#replace all na values -1 at first to indicate they are NaN values
df=df.replace(np.nan, -1)
```

```python
#list of all countries in cvs file
country_list=df['country'].unique().tolist()
country_list

#the list will contain minimum daily_vac number for every country
minimum_list=[]
for country in country_list:
  to_be_add=[]
  for i in df.itertuples(index=False):
    if (i.country==country):
      to_be_add.append(i.daily_vaccinations)
  min_value=min([i for i in to_be_add if i>=0], default="EMPTY")
  if (min_value=='EMPTY'):  #country does not have any valid vaccination number yet, fill it with “0” (zero). 
      min_value=0
  minimum_list.append(min_value)
  
```

```python
#merge countries and their min vaccine num in a dictionary
keys = country_list
values = minimum_list
dictionary = dict(zip(keys, values))
print(dictionary)
```

```python
#replace -1 values with minumum vaccine value of a country 
for index, row in df.iterrows():
  for x in dictionary:
    if (x==row.country and row.daily_vaccinations==-1):
      df.at[index,'daily_vaccinations']=dictionary[x]
df
```

# List the top-3 countries with highest median daily vaccination numbers

```python
median_list=[]
for country in country_list:
  to_be_add=[]
  for i in df.itertuples(index=False):
    if (i.country==country):
      to_be_add.append(i.daily_vaccinations)  #seperate lists for every country, than find median
  median_value=statistics.median(to_be_add)
  median_list.append(median_value)
```

```python
#merge countries and their median vaccine num in a dictionary
keys = country_list
values = median_list
dictionary_median = dict(zip(keys, values))
print(dictionary_median)
```

```python
#sorted list of country-median dictionary
dictionary_median=sorted(dictionary_median.items(), key=lambda x: x[1], reverse=True)
dictionary_median
```
